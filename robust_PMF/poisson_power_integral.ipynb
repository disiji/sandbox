```python
def func_lambda(_lambda):
    beta = 0.1
    X_short = [i+_lambda for i in range(-100,100) if i+_lambda >= 0]
    X_long = [i+_lambda for i in range(-5000,5000) if i+_lambda >= 0]
    val_short = [poisson.pmf(_,_lambda) ** beta for _ in X_short]
    val_long = [poisson.pmf(_,_lambda) ** beta for _ in X_long]
    return sum(val_short)/sum(val_long)
lambda_range = [_ for _ in range(5,10000)]
res_lambda = [func_lambda(_) for _ in lambda_range]
```
