```python
mu_1 = -50
mu_2 = 50
sigma = 10
pi_1 = 0.3
pi_2 = 0.7

x_1 = np.linspace(mu_1-3*sigma, mu_1+3*sigma, 100000)
x_2 = np.linspace(mu_2-3*sigma, mu_2+3*sigma, 100000)
x = np.linspace(mu_1-3*sigma, mu_2+3*sigma, 100000)

plt.style.use('ggplot')
f, axarr = plt.subplots(2, sharex=True,figsize=(10,7))
###################Plot components###################
y1 = norm.pdf(x_1,mu_1,sigma)
y2 = norm.pdf(x_2,mu_2,sigma)
y1_ = norm.pdf(x_1,mu_1,sigma*sqrt(2))
y2_ = norm.pdf(x_2,mu_2,sigma*sqrt(2))

axarr[0].plot(x_1,y1,color ='blue', linewidth=2, linestyle="-",label=r'$g(x)$ components')
axarr[0].plot(x_2,y2,color ='blue', linewidth=2, linestyle="-")
axarr[0].plot(x_1,y1_,color ='red', linewidth=2, linestyle="--",label=r'$h(x)$ components')
axarr[0].plot(x_2,y2_,color ='red', linewidth=2, linestyle="--")
axarr[0].text(x=mu_1-3, y=-0.004, s= "$\mu_1$",fontsize=15)
axarr[0].text(x=mu_2-3, y=-0.004, s= "$\mu_2$",fontsize=15)
axarr[0].plot([mu_1,mu_1],[0,norm.pdf(mu_1,mu_1,sigma)],color ='grey', linewidth=1.5, linestyle="--")
axarr[0].plot([mu_2,mu_2],[0,norm.pdf(mu_2,mu_2,sigma)],color ='grey', linewidth=1.5, linestyle="--")
axarr[0].legend(loc='upper left',fontsize = 'large')

axarr[0].set_title('componets of $g(x)$ and $h(x)$')
axarr[0].set_xlim([mu_1 - 10*sigma,mu_2+4*sigma])
axarr[0].set_yticklabels([])
###################Plot pdf###################
mu_kl = pi_1 * mu_1 + pi_2 * mu_2
mu_l2 = mu_2

g_x = pi_1*norm.pdf(x,mu_1,sigma) + pi_2*norm.pdf(x,mu_2,sigma)
h_x = pi_1*norm.pdf(x,mu_1,sigma*sqrt(2)) + pi_2*norm.pdf(x,mu_2,sigma*sqrt(2))
f_kl_x = norm.pdf(x,mu_kl,sigma)
f_l2_x = norm.pdf(x,mu_l2,sigma)

axarr[1].plot([mu_kl], [0], marker='*', markersize=10, color="black")
axarr[1].plot([mu_2], [0], marker='*', markersize=10, color="black")
axarr[1].text(x=mu_kl-3, y=-0.006, s= "$\mu_{KL}$",fontsize=15)
axarr[1].text(x=mu_2-3, y=-0.006, s= "$\mu_{L_2E}$",fontsize=15)

axarr[1].plot(x,g_x,color ='blue', linewidth=2, linestyle="-",label=r'$g(x)$ density')
axarr[1].plot(x,h_x,color ='red', linewidth=2, linestyle="--",label=r'$h(x)$ density')
axarr[1].plot(x,f_kl_x,color ='black', linewidth=2.5, linestyle= ':' ,label=r'$f_{KL}(x)$ density')
axarr[1].plot(x,f_l2_x,color ='black', linewidth=2.5, linestyle="-.",label=r'$f_{L_2}(x)$ density')
# axarr[1].text(x=mu_1-3, y=-0.004, s= "$\mu_1$",fontsize=15)
# axarr[1].text(x=mu_2-3, y=-0.004, s= "$\mu_2$",fontsize=15)
axarr[1].plot([mu_1,mu_1],[0,pi_1*norm.pdf(mu_1,mu_1,sigma)],color ='grey', linewidth=1.5, linestyle="--")
axarr[1].plot([mu_2,mu_2],[0,pi_2*norm.pdf(mu_2,mu_2,sigma)],color ='grey', linewidth=1.5, linestyle="--")
axarr[1].plot([mu_kl,mu_kl],[0,norm.pdf(mu_kl,mu_kl,sigma)],color ='grey', linewidth=1.5, linestyle="--")
axarr[1].legend(loc='upper left',fontsize = 'large')

axarr[1].set_title('density of $g(x)$,  $h(x)$, and $f(x)$')
axarr[1].set_xlim([mu_1 - 10*sigma,mu_2+4*sigma])

axarr[1].set_xticklabels([])
axarr[1].set_yticklabels([])
```
